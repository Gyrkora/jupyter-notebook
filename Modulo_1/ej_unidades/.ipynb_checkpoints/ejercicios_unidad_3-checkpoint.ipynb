```python
""" Ejercicio 5

Suponga que tiene una verdulería y carga la cantidad y el precio de lo comprado por un cliente. Realice un programa que tome de a uno la cantidad 
y el precio comprado por el cliente y al finalizar la compra retorne el monto total gastado

 """


def calcular_total_compra():
    total = 0
    
    while True:
        try:
            cantidad = int(input("Ingrese la cantidad comprada (0 para finalizar): "))
            if cantidad == 0:
                break
            
            precio = float(input("Ingrese el precio unitario: "))
            
            subtotal = cantidad * precio
            total += subtotal
            
            print(f"Subtotal: {subtotal:.2f}")
            
        except ValueError:
            print("Ingrese valores numéricos válidos.")
    
    return total

total_compra = calcular_total_compra()
print(f"Monto total gastado: {total_compra:.2f}")



# ------------------------------------------------------------------------

# compras = {
#     "aceite": 500,
#     "arroz": 300,
#     "avena": 1200
# }

# total = 0
# for value in compras.values():
#         total += value


# print("el total de las compras son: ",total)


# https://blog.finxter.com/5-best-ways-to-sum-dictionary-values-in-python/
```

```python
"""
Ejercicio 5
Suponga que tiene una verdulería y carga la cantidad y el precio
de lo comprado por un cliente. Realice un programa que tome
de a uno la cantidad y el precio comprado por el cliente y
al finalizar la compra retorne el monto total gastado. 
"""
total=0
valor=False
eleccion=input("Para iniciar ingrese 'i', para finalizar ingrese otra tecla: ")
if eleccion=='i':
    valor=True
else:
    valor=False

while valor==True:
    cantidad=float(input("Ingrese la cantidad: "))
    precio=float(input("Ingrese el precio: "))
    total=total+cantidad*precio
    
    eleccion=input("Para iniciar ingrese 'i', para finalizar ingrese otra tecla: ")

    if eleccion=='i':
        valor=True
    else:
        valor=False


print("El valor total de la compra es: ", total)
```

```python

""" Ejercicio 6

A partir del ejercicio 5 cree un programa que vaya agregando en una lsta las compras realizadas

 """


def calcular_total_compra():
    total = 0
    lista_de_compras = []
    
    while True:
        try:
            cantidad = int(input("Ingrese la cantidad comprada (0 para finalizar): "))
            if cantidad == 0:
                break
            
            precio = float(input("Ingrese el precio unitario: "))
            
            subtotal = cantidad * precio
            total += subtotal
            lista_de_compras.extend([subtotal])
            print("esta es la lista de compras ", lista_de_compras)
            
            print(f"Subtotal: {subtotal:.2f}")
            
        except ValueError:
            print("Ingrese valores numéricos válidos.")
    
    return total

total_compra = calcular_total_compra()
print(f"Monto total gastado: {total_compra:.2f}")


# --------------------------------------------------

# compras = {
#     "aceite": 500,
#     "arroz": 300,
#     "avena": 1200
# }

# list_de_compras = []

# for key in compras:
#     list_de_compras.append((key))


# print("la lista de compras es: ",list_de_compras)
```
