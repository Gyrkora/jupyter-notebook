<h2>Tipos de relaciones</h2>


<h3>Relación uno a uno (1:1) </h3>

<p>Se produce cuando la relación se realiza solo entre un registro de una tabla con un registro de otra. 
Es una de las relaciones más utilizadas, ya que permiten una relación de tipo exclusivo. En esta relación, los campos establecidos como primary key de ambas tablas están enlazados.Algunos ejemplos de este tipo de relaciones en bases de datos 
1:1 lo encontramos en la asignación de banderas por país, donde cada bandera solo se puede asignar a un único país. Lo mismo ocurre con otros datos como las matrículas de coche (cada coche solo puede tener un único número de matrícula, y cada matrícula solo puede corresponder a un coche en concreto) o el número de serie de un producto (cada producto tiene un número único de serie, y cada número de serie corresponde a un único producto).
</p>

<h3>Relación uno a varios (1:N)</h3>

<p>En este tipo de relación de uno a varios, un registro de una tabla puede enlazarse a varios registros de otra tabla. La primera key está vinculada a varios registros de otra tabla. Esta relación es la más utilizada en las bases de datos relacionales. 

Veamos ejemplos de este tipo de relación 1:N:

Dirección postal que puede relacionarse con varias personas que pueden vivir en el mismo.
Nombre de una empresa que puede relacionarse con diversos trabajadores de la misma.
Nombre de cliente que puede tener distintos pedidos de venta

Ejemplo:




<h3>Relaciones varios a varios (N:N)</h3>

<p>Cuando varios registros de una tabla pueden relacionarse con varios registros de otra tabla se está ante una relación del tipo varios a varios. Este tipo de relación es la menos habitual en una base de datos relacional.

Un ejemplo de este tipo de tablas N:N lo podemos ver con una tabla de equipo de fútbol y otra de jugadores. Se puede dar el caso de que en un equipo de fútbol juegan varios jugadores y, que además, un jugador haya jugado en varios equipos de fútbol.

Otro ejemplo lo podemos encontrar en una base de datos con una tabla de asignaturas y otra de profesores. Varios profesores pueden dar una misma asignatura, así como un profesor puede impartir varias asignatu.

<h2>Acciones en SQLite usando SQLite Studio</h2>


<h3>Crear base y tabla</h3>


```python
import sqlite3


def crear_base():
    con = sqlite3.connect('mibase.db')
    return con


def crear_table(con):
    cursor = con.cursor()
    sql = "CREATE TABLE alimento(id integer PRIMARY KEY, nombre)"
    cursor.execute(sql)
    con.commit()


con = crear_base()
crear_table(con)

# close db
# con.close() 
```

<h3>Insertar</h3>


```python
import sqlite3

def crear_base():
    con = sqlite3.connect("mibase.db")
    return con

def insertar(con, my_id, nombre):
    cursor = con.cursor()
    my_id = int(my_id)
    data = (my_id, nombre)
    sql = "INSERT INTO alimento(id, nombre) VALUES(?, ?)"
    cursor.execute(sql, data)
    con.commit()

con = crear_base()
insertar(con,4,  "Pera")
insertar(con,5,  "Durazno")
```

<h3>Borrar</h3>


```python
import sqlite3

def crear_base():
    con = sqlite3.connect("mibase.db")
    return con

def borrar(con, my_id):
    cursor = con.cursor()
    my_id = int(my_id)
    data = (my_id, )
    sql = "DELETE FROM alimento WHERE id = ?;"
    cursor.execute(sql, data)
    con.commit()

con = crear_base()
borrar(con, 2)
```

<h3>Actualizar</h3>


```python
import sqlite3

def crear_base():
    con = sqlite3.connect("mibase.db")
    return con

def update(con, my_id, nombre):
    cursor = con.cursor()
    my_id = int(my_id)
    data = (nombre, my_id)
    sql = "UPDATE alimento SET nombre=? WHERE id=?"
    cursor.execute(sql, data)
    con.commit()

con = crear_base()
update(con,3,  "Uvas")
```

<h3>Seleccionar</h3>


```python
import sqlite3

def crear_base():
    con = sqlite3.connect("mibase.db")
    return con

def seleccionar(con, my_id):
    cursor = con.cursor()
    my_id = int(my_id)
    data = (my_id, )
    sql = "SELECT * FROM alimento WHERE id=?"
    cursor.execute(sql, data)
    con.commit()

    rows = cursor.fetchall()

    for row in rows:
        print(row) 

con = crear_base()
seleccionar(con,4)
```

<h3>Shelve</h3>

```python
# insert

import shelve


gyr = {"nombre": "Gyrkora", "edad": 31}
alisa = {"nombre": "Alisa", "edad": 30}

# si no existe, la crea
db = shelve.open('persona')
db['gyr'] = gyr
db['alisa'] = alisa
db.close()
```

```python
# read

import shelve

db = shelve.open('persona')
for key in db:
    print(key, '=>', db[key])
    print(key, '=>' , db[key]['nombre'], db[key]['edad'])
db.close()
```

```python
# delete

import shelve

db = shelve.open('persona')

del db['gyr']
db.close()
```

```python
# update

import shelve

db = shelve.open('persona')
gyr = db['gyr']
gyr = {'nombre': 'gyr', "edad": 80}
db['gyr']=gyr
db.close()
```
