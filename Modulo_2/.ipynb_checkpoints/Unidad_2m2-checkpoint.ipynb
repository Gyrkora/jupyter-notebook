<h2>Tipos de relaciones</h2>


<h3>Relacin uno a uno (1:1) </h3>

<p>Se produce cuando la relacin se realiza solo entre un registro de una tabla con un registro de otra. 
Es una de las relaciones ms utilizadas, ya que permiten una relacin de tipo exclusivo. En esta relacin, los campos establecidos como primary key de ambas tablas estn enlazados.Algunos ejemplos de este tipo de relaciones en bases de datos 
1:1 lo encontramos en la asignacin de banderas por pas, donde cada bandera solo se puede asignar a un nico pas. Lo mismo ocurre con otros datos como las matrculas de coche (cada coche solo puede tener un nico nmero de matrcula, y cada matrcula solo puede corresponder a un coche en concreto) o el nmero de serie de un producto (cada producto tiene un nmero nico de serie, y cada nmero de serie corresponde a un nico producto).
</p>

<h3>Relacin uno a varios (1:N)</h3>

<p>En este tipo de relacin de uno a varios, un registro de una tabla puede enlazarse a varios registros de otra tabla. La primera key est vinculada a varios registros de otra tabla. Esta relacin es la ms utilizada en las bases de datos relacionales. 

Veamos ejemplos de este tipo de relacin 1:N:

Direccin postal que puede relacionarse con varias personas que pueden vivir en el mismo.
Nombre de una empresa que puede relacionarse con diversos trabajadores de la misma.
Nombre de cliente que puede tener distintos pedidos de venta

Ejemplo:




<h3>Relaciones varios a varios (N:N)</h3>

<p>Cuando varios registros de una tabla pueden relacionarse con varios registros de otra tabla se est ante una relacin del tipo varios a varios. Este tipo de relacin es la menos habitual en una base de datos relacional.

Un ejemplo de este tipo de tablas N:N lo podemos ver con una tabla de equipo de ftbol y otra de jugadores. Se puede dar el caso de que en un equipo de ftbol juegan varios jugadores y, que adems, un jugador haya jugado en varios equipos de ftbol.

Otro ejemplo lo podemos encontrar en una base de datos con una tabla de asignaturas y otra de profesores. Varios profesores pueden dar una misma asignatura, as como un profesor puede impartir varias asignatu.

<h2>Acciones en SQLite usando SQLite Studio</h2>


<h3>Crear base y tabla</h3>


```python
import sqlite3


def crear_base():
    con = sqlite3.connect('mibase.db')
    return con


def crear_table(con):
    cursor = con.cursor()
    sql = "CREATE TABLE alimento(id integer PRIMARY KEY, nombre)"
    cursor.execute(sql)
    con.commit()


con = crear_base()
crear_table(con)

# close db
# con.close() 
```

<h3>Insertar</h3>


```python
import sqlite3

def crear_base():
    con = sqlite3.connect("mibase.db")
    return con

def insertar(con, my_id, nombre):
    cursor = con.cursor()
    my_id = int(my_id)
    data = (my_id, nombre)
    sql = "INSERT INTO alimento(id, nombre) VALUES(?, ?)"
    cursor.execute(sql, data)
    con.commit()

con = crear_base()
insertar(con,4,  "Pera")
insertar(con,5,  "Durazno")
```

<h3>Borrar</h3>


```python
import sqlite3

def crear_base():
    con = sqlite3.connect("mibase.db")
    return con

def borrar(con, my_id):
    cursor = con.cursor()
    my_id = int(my_id)
    data = (my_id, )
    sql = "DELETE FROM alimento WHERE id = ?;"
    cursor.execute(sql, data)
    con.commit()

con = crear_base()
borrar(con, 2)
```

<h3>Actualizar</h3>


```python
import sqlite3

def crear_base():
    con = sqlite3.connect("mibase.db")
    return con

def update(con, my_id, nombre):
    cursor = con.cursor()
    my_id = int(my_id)
    data = (nombre, my_id)
    sql = "UPDATE alimento SET nombre=? WHERE id=?"
    cursor.execute(sql, data)
    con.commit()

con = crear_base()
update(con,3,  "Uvas")
```

<h3>Seleccionar</h3>


```python
import sqlite3

def crear_base():
    con = sqlite3.connect("mibase.db")
    return con

def seleccionar(con, my_id):
    cursor = con.cursor()
    my_id = int(my_id)
    data = (my_id, )
    sql = "SELECT * FROM alimento WHERE id=?"
    cursor.execute(sql, data)
    con.commit()

    rows = cursor.fetchall()

    for row in rows:
        print(row) 

con = crear_base()
seleccionar(con,4)
```

<h3>Shelve</h3>

```python
# insert

import shelve


gyr = {"nombre": "Gyrkora", "edad": 31}
alisa = {"nombre": "Alisa", "edad": 30}

# si no existe, la crea
db = shelve.open('persona')
db['gyr'] = gyr
db['alisa'] = alisa
db.close()
```

```python
# read

import shelve

db = shelve.open('persona')
for key in db:
    print(key, '=>', db[key])
    print(key, '=>' , db[key]['nombre'], db[key]['edad'])
db.close()
```

```python
# delete

import shelve

db = shelve.open('persona')

del db['gyr']
db.close()
```

```python
# update

import shelve

db = shelve.open('persona')
gyr = db['gyr']
gyr = {'nombre': 'gyr', "edad": 80}
db['gyr']=gyr
db.close()
```
