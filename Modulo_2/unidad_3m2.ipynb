<h4>POSIX, PCRE y FLAGS</h4>

<p><strong>Segn CHATGPT</strong></p>

<p>a) **POSIX** (Portable Operating System Interface) es un conjunto de estndares que especifica interfaces para sistemas operativos tipo Unix. Fue desarrollado para promover la portabilidad de las aplicaciones entre diferentes variantes de sistemas operativos Unix y sistemas similares. Los estndares POSIX definen cmo deben comportarse las API (Interfaz de Programacin de Aplicaciones) y las bibliotecas del sistema en estas plataformas. Esto ayuda a los desarrolladores a escribir cdigo que sea compatible con varios sistemas Unix y Unix-like, como Linux, macOS, y varios sistemas BSD.

b) **PCRE** (Perl Compatible Regular Expressions) es una biblioteca de expresiones regulares desarrollada en base a la sintaxis de las expresiones regulares utilizadas en el lenguaje de programacin Perl. PCRE proporciona una manera poderosa de buscar y manipular cadenas de texto utilizando patrones de bsqueda flexibles y avanzados. Esta biblioteca se utiliza ampliamente en diversos lenguajes de programacin, como Python, PHP y Ruby, para realizar operaciones de bsqueda y manipulacin de cadenas.

Diferencias esenciales:

- **POSIX** es un conjunto de estndares para sistemas operativos tipo Unix, mientras que **PCRE** es una biblioteca de expresiones regulares utilizada para buscar y manipular cadenas de texto.

- **POSIX** no est relacionado directamente con expresiones regulares; en cambio, se centra en la interfaz del sistema operativo. **PCRE** se enfoca especficamente en expresiones regulares.

- **POSIX** se aplica a la programacin de sistemas y define estndares para APIs, bibliotecas y comandos del sistema. **PCRE** se utiliza en programacin de aplicaciones para realizar bsquedas y manipulaciones de patrones en cadenas de texto.

c) **Flags** en el contexto de programacin generalmente se refieren a marcadores o indicadores que se utilizan para modificar el comportamiento de una funcin o mtodo. Estos indicadores se pasan como argumentos a una funcin o se establecen como atributos en objetos para controlar ciertas acciones o configuraciones.

En el contexto de las expresiones regulares (regex), las **flags** son marcadores que se utilizan para modificar cmo se realiza la coincidencia de patrones. Algunos ejemplos comunes de flags en expresiones regulares incluyen:

- **i**: Realizar una bsqueda insensible a maysculas y minsculas.
- **g**: Realizar una bsqueda global para encontrar todas las coincidencias en lugar de detenerse despus de la primera coincidencia.
- **m**: Habilitar el modo multilnea, lo que significa que "^" y "$" coincidirn con el inicio y el final de cada lnea en lugar del inicio y el final del texto completo.

En resumen, los **flags** en el contexto de programacin son indicadores utilizados para controlar el comportamiento de una funcin o mtodo, mientras que en el contexto de las expresiones regulares, son marcadores utilizados para modificar cmo se realiza la coincidencia de patrones en una 

cadena de texto.</p>

```python
import re

patron = re.compile(r'(?i:pera) y manzana')

ej1 = "pera y manzana"
ej2 = "Pera y manzana"
ej3 = "PerA Y manzana"

print(patron.match(ej1))
print(patron.match(ej2))
print(patron.match(ej3))
```

```python
import re

patron = re.compile(r"color (?i:(?=rojo))")
string1 = "El auto de color rojo"
string2 = "El auto de color RoJo"
print(patron.search(string1))
print(patron.search(string2))
```

```python
import re

# el patrn ((?P<name>ap8)(?(name)9|) ) busca coincidencias que comiencen con "ap8" y que estn seguidas por "9" si el grupo de captura "name" existe en la coincidencia anterior
patron1 = re.compile(r"((?P<name>ap8)(?(name)9|) )")
mail = "ap89 ap8 rp8 rp4 w"
print(patron1.search(mail))
```

```python
import re

patron1 = re.compile(r"(<)?(\w+@\w+(?:\.[a-z]+)+)(?(1)>|$)", re.I )

mail = "<user@host.Com>"
print(patron1.search(mail))
```

```python
import re

retorno1 = re.split(r"\W+", "Palabras, palabras, palabras.")
retorno2 = re.split(r"(\W+)", "Palabras, palabras, palabras.")
retorno3 = re.split(r"\W+", "Palabras, palabras, palabras.", 1)
retorno4 = re.split("[a-f]+", "0a3B8", flags=re.IGNORECASE)

print(retorno1)
print(retorno2)
print(retorno3)
print(retorno4)
```

```python
# uso del ^ 

import re

# Buscar caracteres que no sean vocales
texto = "Hola, mundo!"
patron = re.compile(r'[^aeiou]')
coincidencias = patron.findall(texto)
print(coincidencias)
```
