<h3>Como empezar un proyecto?</h3>


```python
from tkinter import *

#campos de entrada de texto
root = Tk()
e = Entry(root)
e.pack()


""" Python """

a = 4
b = 5
c = a + b
d = f"mi resultado es {str(c)}"


""" Python  """
#tipo de dato entero de tkinter
# var = IntVar() 
var = StringVar()

e.config(textvariable=var)
var.set(d)
root.mainloop()
```

```python
from tkinter import *

# ventana principal / solo una por app
root = Tk() 

# widget donde agregar un nombre. Dentro de qué elemento se agrega, es decir, siempre irá dentro de la ventana y ahí le pondré el texto

""" setting the names of the columns and rows """

# what is it - it is a text that says "id"
id_row = Label(root, text="id", fg="red")
# where it is
id_row.grid(row=0, column=0, sticky=W)
name_row = Label(root, text="Nombre")
name_row.grid(row=1, column=0, sticky=W)

# campos de entrada
entry_id = Entry(root)
entry_id.grid(row=0, column=1)
entry_name = Entry(root)
entry_name.grid(row=1, column=1)

def save_function():print("hola")

# boton para guardar los datos en la base de datos
button_send = Button(root, text="Guardar", command=save_function)
button_send.grid(row=2, column=1)


#se "cierra" la ventana
root.mainloop()
```

<h3>¿Cómo agregar un TreeView?</h3>

```python
from tkinter import *
from tkinter import ttk


my_id = 0

# ventana principal / solo una por app
root = Tk() 

var_nombre = StringVar()
var_apellido = StringVar()

root.geometry("210x300") # giving the window its height and width
root.config(background = "black") 


""" setting the names of the columns and rows """

nombre = Label(root, text="Nombre", fg="white", bg="black" )
nombre.grid(row=0, column=0, sticky=W)
apellido = Label(root, text="Apellido", fg="white", bg="black")
apellido.grid(row=1, column=0, sticky=W)

# campos de entrada

entry_nombre = Entry(root, textvariable=var_nombre, width=25) #una variable que se asocia

entry_nombre.grid(row=0, column=1)
entry_apellido = Entry(root, textvariable=var_apellido, width=25) #accedo a las variables
entry_apellido.grid(row=1, column=1)



""" setting functions """

def fn_g():
    global my_id
    my_id += 1
    tree.insert("", "end", text=str(my_id), values=(var_nombre.get(), var_apellido.get() ))

def fn_b():
    global my_id
    item = tree.focus()
    print(item)
    print(tree.item(item))
    tree.delete(item)
    my_id -= 1


""" Setting the tree """

tree = ttk.Treeview(root)
tree["columns"] = ("col1", "col2")
tree.column("#0", width=50, minwidth=50, anchor=W)
tree.column("col1", width=80, minwidth=80, anchor=W)
tree.column("col2", width=60, minwidth=80, anchor=W)


#adding it to the window
tree.grid(column=0, row=4, columnspan=4)


""" buttons """

boton_g = Button(root, text="Guardar", command=fn_g)
boton_g.grid(row=2, column=1)
boton_b = Button(root, text="Eliminar", command=fn_b)
boton_b.grid(row=3, column=1)
print(boton_b.winfo_class())




root.mainloop()
```

<h3>Mensajes</h3>

```python
from tkinter import *
from tkinter.messagebox import *

root = Tk()

def fn_g():
    showerror("error message title", "message content")
def fn_s():
    if askyesno("titulo consulta de verificación", "contenido de la consulta"):
        showinfo("consulta exitosa", "la consulta se realizó correctamente")
    else:
        showinfo("No", "Esta a punto de salir")

btn_e = Button(root, text="Error", command=fn_g)
btn_e.grid(row=2, column=1)


btn_s = Button(root, text="Show", command=fn_s)
btn_s.grid(row=3, column=1)

root.mainloop()
```

<h3>¿Cómo agrego o abro imágenes?</h3>

```python
# ejemplo del profe

from tkinter import *
import tkinter as tk
from PIL import ImageTk, Image
import os

BASE_DIR = os.path.dirname((os.path.abspath(__file__)))
ruta = os.path.join(BASE_DIR, "img", "bg_mujeres.jpeg")

""" Formas de posicionar en tkinter

grid
pack
place

 """

window = Tk()
window.title("Imagen con PIL")
image2 = Image.open(ruta)
image1 = ImageTk.PhotoImage(image2)
background_label = tk.Label(window, image=image1)
background_label.place(x=0, y=0, relwidth=1, relheight=1)
window.mainloop
```

<h3>¿Cómo hago menúes en Tkinter?</h3>

```python
from tkinter import *

root = Tk()

def fn():
    print("Hello")

""" Menu principal """

menu_bar = Menu(root)

#Archivo - 1st column
menu_archivo = Menu(menu_bar, tearoff=0)
menu_archivo.add_command(label="Abrir", command=fn)
menu_archivo.add_command(label="Guardar", command=fn)
menu_archivo.add_separator()
menu_archivo.add_command(label="Salir", command=root.quit)
menu_bar.add_cascade(label="Archivo", menu=menu_archivo)

#Editar - 2nd column 
menu_edit = Menu(menu_bar, tearoff=1)
menu_edit.add_command(label="Cortar", command=fn)
menu_edit.add_command(label="Preferencias", command=fn)
menu_bar.add_cascade(label="Editar", menu=menu_edit)

""" submenu """

submenu = Menu(menu_edit, tearoff=True)
submenu.add_command(label="Rotar", command=fn)
submenu.add_separator()
submenu.add_command(label="Eliminar", command=fn)
menu_edit.add_cascade(label="Más", menu=submenu)



root.config(menu=menu_bar)
root.mainloop()
```
