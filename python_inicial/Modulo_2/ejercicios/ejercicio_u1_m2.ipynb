```python
""" Ejercicio único

Realizar una app utilizando Tkinter
Características: la misma debe presentar dos botones

- El primero de alta de datos, debe permitir tomar los valores de los campos de entrada y presentarlos en la consola 

- El segundo debe permitir cambiar el color de fondo de la app

"""

from tkinter import *
from tkinter import ttk
from tkinter.colorchooser import askcolor  



##--Modelo--##

def fn_alta():
    print(f"alimento: {var_nombre.get()}, tipo: {var_tipo.get()}")

def fn_color_change():
    color_picked = askcolor(color="#00ff00", title="title")
    root.config(background=color_picked[1])

##--Vista--##

""" global """

root = Tk()

var_nombre = StringVar()
var_tipo = StringVar()
bg_color = StringVar()

root.geometry("290x150") 
root['pady'] = 20
root.title("Tipos de alimentos")
root.config(background = "black") 


""" entries """

nombre = Label(root, text="Nombre de alimento", fg="white", bg="black" )
nombre.grid(row=0, column=0, sticky=W)
tipo = Label(root, text="tipo de alimento", fg="white", bg="black")
tipo.grid(row=1, column=0, sticky=W)
entry_nombre = Entry(root, textvariable=var_nombre, width=25)
entry_nombre.grid(row=0, column=1)
entry_tipo = Entry(root, textvariable=var_tipo, width=25)

entry_tipo.grid(row=1, column=1)

""" buttons """

btn_alta = Button(root, text="Guardar", command=fn_alta)
btn_alta.grid(row=3, column=1)
btn_color = Button(root, text="Color change", command=fn_color_change)
btn_color.grid(row=4, column=1)


root.mainloop()
```
