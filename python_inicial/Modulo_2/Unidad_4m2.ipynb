<h3>Parametrize</h3>

<p>The builtin pytest.mark.parametrize decorator enables parametrization of arguments for a test function. Here is a typical example of a test function that implements checking that a certain input leads to an expected output:

</p>

```python
# python code

from operaciones import sumar
from operaciones import multiplicar
import pytest



@pytest.mark.parametrize(
    "a , b, resultado",
    [
        (1, 2, 2),
        (sumar(1, 2), 2, 6),
        (5, 2, 10),
        (multiplicar(2, 2), 2, 8),
    ]
)

def test_multiplicar(a, b, resultado):
    assert multiplicar(a, b,) == resultado
```

<h3>Fixture</h3>

<p>Software test fixtures initialize test functions. They provide a fixed baseline so that tests execute reliably and produce consistent, repeatable, results. Initialization may setup services, state, or other operating environments. These are accessed by test functions through arguments; for each fixture used by a test function there is typically a parameter (named after the fixture) in the test functions definition.

pytest fixtures offer dramatic improvements over the classic xUnit style of setup/teardown functions:

fixtures have explicit names and are activated by declaring their use from test functions, modules, classes or whole projects.

fixtures are implemented in a modular manner, as each fixture name triggers a fixture function which can itself use other fixtures.

fixture management scales from simple unit to complex functional testing, allowing to parametrize fixtures and tests according to configuration and component options, or to re-use fixtures across function, class, module or whole test session scopes.

teardown logic can be easily, and safely managed, no matter how many fixtures are used, without the need to carefully handle errors by hand or micromanage the order that cleanup steps are added.</p>


```python
from operaciones import sumar
from operaciones import multiplicar
import pytest

@pytest.fixture
def fixture_1():
    print("desde fixture")
    return 5

def test_sumar_f(fixture_1):
    print("print desde test")
    variable = fixture_1
    assert sumar(variable, 2)==8
```

<h2>Usando yield</h2>

```python
from operaciones import sumar
from operaciones import multiplicar
import pytest

@pytest.fixture
def fixture_1():
    print("desde fixture antes")
    yield 5
    print("desde fixture despus")


def test_sumar_f(fixture_1):
    print("print desde test")
    variable = fixture_1
    assert sumar(variable, 2)==8
```
