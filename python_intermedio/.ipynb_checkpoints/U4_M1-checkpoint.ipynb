<strong style="color: orange;">Práctica de clases</strong>


```python
# main

if __name__ == "__main__":
    from classes.cls_personas import Personas
    from classes.cls_gerentes import Gerentes


    susana = Personas("Susana Palacios", 300)
    carlos = Personas("Carlos Jimnez", 400)
    susana.dar_aumento(0.5)
    carlos.dar_aumento(0.2)
    tom=Gerentes("Tomas Perez", 500)
    tom.dar_aumento(0.1)

    base_personas = [susana, carlos, tom]
    for persona in base_personas:
        print(persona.apellido(), persona.salario)
```

```python
# class personas

class Personas():
    def __init__(self, nombre, salario):
        self.nombre = nombre
        self.salario = salario

    def apellido(self,):
        return self.nombre.split()[-1]
    

    def dar_aumento(self, porcentaje):
        self.salario *= 1 + porcentaje
```

```python
# class gerente

from classes.cls_personas import Personas


class Gerentes(Personas):


    def dar_aumento(self, porcentaje, premio=0.1):
        self.salario *= 1 + porcentaje + premio
    
```

<strong style="color: orange;">Creación de botón</strong>

```python
# btn base

from tkinter import *
from params import bcolor, bfont, bsize

class BotonB:
    def __init__(self, parent=None, *args ,**kwars):
        self.root = parent
        self.root.geometry("300x300")
        b1 = Button(self.root, *args, **kwars)
        b1.pack(side=LEFT)
        b1.config(command=self.callback, bg=bcolor, font=(bfont, bsize))

    def callback(self,):
            print('funcionando')
            self.root.quit()


if __name__=="__main__":
    root=Tk()
    mi_app = BotonB(root, text="soy el boton base")
    root.mainloop()
```

```python
# btn secundario

from tkinter import *
from btn_base import BotonB

class MiBoton(BotonB):        
    def callback(self,):
            print('estoy en el boton secundario')


if __name__=="__main__":
    root=Tk()
    mi_app = MiBoton(root, text="btn secundario")
    root.mainloop()
```

```python
# params de estilos

bcolor = "yellow"
bfont = "ARIAL"
bsize = 10
bstyle = "bold"
bborder = "solid"
bgap = 0.5
```
