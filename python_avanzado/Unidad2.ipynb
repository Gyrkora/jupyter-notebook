````python
# one way


# Python program to illustrate property() function
class Person:
    def __init__(self, name):
        self._name = name

    # getter function
    def get_name(self):
        print('Getting name...')
        return self._name

    # setter function
    def set_name(self, value):
        print('Setting name to:', value)
        self._name = value

    # deleter function
    def del_name(self):
        print('Deleting name...')
        del self._name

    # Set property() to use get_name, set_name and del_name methods
    name = property(get_name, set_name, del_name)

p = Person('David')
print(p.name)
p.name = 'Rocky'
del p.name
````

````python
# better way


# Python program to illustrate @property decorator
class Person:
    def __init__(self, name):
        self._name = name

    @property
    def name(self):
        print('Getting name...')
        return self._name

    @name.setter
    def name(self, value):
        print('Setting name to:', value)
        self._name = value

    @name.deleter
    def name(self):
        print('Deleting name...')
        del self._name

p = Person('David')
print(p.name)
p.name = 'Rocky'
del p.name
````

We have used the @property decorator in this program instead of directly using the property() function. First, we specify the attribute name() method for the class Person. We do this by using the @property decorator before our getter function. We then use the same attribute name to define the setter and deleter. This can be seen as @name.setter and @name.deleter. When we use p.name now, it calls the necessary getter, setter, and deleter, as evidenced by the displayed output inside the method.

### Descriptores Class:

```python
class Descriptores:
    def __get__(self, instance, owner):
        print('Recuperar el usuario...')
        return instance._usuario.upper()

    def __set__(self, instance, valor):
        print('Modificar usuario...')
        instance._usuario = valor

    def __delete__(self, instance):
        print('Remover el usuario....')
        del instance._usuario
```

In this descriptor class:

- **`__get__(self, instance, owner)`**: This method is called when the attribute is accessed. `instance` is the instance of the owning class (`Cliente` in this case), and `owner` is the class itself. It returns the uppercase version of the `_usuario` attribute.

- **`__set__(self, instance, valor)`**: This method is called when the attribute is set. It prints a message and sets the `_usuario` attribute to the given value.

- **`__delete__(self, instance)`**: This method is called when the attribute is deleted. It prints a message and deletes the `_usuario` attribute.

### Cliente Class:

```python
class Cliente:
    def __init__(self, usuario):
        self._usuario = usuario
    
    usuario = Descriptores()
```

In this class:

- **`usuario = Descriptores()`**: Here, the `usuario` attribute is an instance of the `Descriptores` class. This means that the descriptor's behavior will be applied when accessing, setting, or deleting the `usuario` attribute.

### Example Usage:

```python
cliente1 = Cliente("Anna")
cliente1.usuario = "Carlos"  # Calls the __set__ method of Descriptores
print(cliente1.usuario)      # Calls the __get__ method of Descriptores
```

In this example:

1. An instance of the `Cliente` class is created with the username "Anna."
2. Setting `cliente1.usuario = "Carlos"` calls the `__set__` method of `Descriptores`, which prints "Modificar usuario..." and sets `_usuario` to "Carlos."
3. Accessing `print(cliente1.usuario)` calls the `__get__` method of `Descriptores`, which prints "Recuperar el usuario..." and returns the uppercase version of "Carlos."

### Additional Example:

Let's extend the example with a descriptor that checks the length of the username:

class LengthValidator:

```python

```

This demonstrates how descriptors can be used to enforce specific behavior and constraints when getting and setting attributes.
