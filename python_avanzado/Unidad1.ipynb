
## Delegación

La delegación en programación se refiere a la práctica de pasar la responsabilidad de realizar una tarea a otro objeto. En Python, esto se puede lograr mediante la composición, donde una clase utiliza instancias de otras clases para realizar ciertas operaciones en lugar de realizarlas directamente. Esto fomenta la reutilización de código y permite una mayor flexibilidadon:

```python
class Personas: 
    def manejar(self,):
        self.accion()

class Ingleses(Personas):
    def accion(self):
        print("los ingleses conducen a la izquierda")

x = Ingleses()
x.manejar()
```

```python
class Motor:
    def encender(self):
        print("Motor encendido")

    def apagar(self):
        print("Motor apagado")


class Auto:
    def __init__(self):
        # Delegación: Creamos una instancia de la clase Motor
        self.motor = Motor()

    def arrancar(self):
        print("Arrancando el auto")
        # Delegación: Llamamos al método encender del objeto Motor
        self.motor.encender()

    def apagar(self):
        print("Apagando el auto")
        # Delegación: Llamamos al método apagar del objeto Motor
        self.motor.apagar()


# Uso del ejemplo
mi_auto = Auto()

mi_auto.arrancar()
mi_auto.apagar()
```

En este ejemplo, la clase Auto tiene un atributo motor, que es una instancia de la clase Motor. Cuando se llama al método arrancar en un objeto Auto, este método delega la tarea de encender el motor al objeto Motor. Similarmente, cuando se llama al método apagar, delega la tarea de apagar el motor al objeto Motor.

La delegación permite una estructura más modular y fácil de entender, ya que cada clase tiene responsabilidades específicas y no realiza tareas fuera de su ámbito.

## Dunder or Magic Methods in Python

### "__ str __"

```python
class Usuarios:
    def __init__(self, nombre):
        self.nombre = nombre
    def __str__(self) -> str: 
        return "El nombre de usuario es: " + self.nombre
    

x = Usuarios("Anna")
print(x)
```

### "__ getitem __" + " __ setitem __"

```python


class Indice:

    lista=["m", "e", "l", "o", "n"]
    def __getitem__(self, mi_indice):
        return mi_indice*2
    
    def __setitem__(self, mi_indice, valor):
        self.lista[mi_indice] = valor
    

x = Indice()
print(x[3])  # imprime 6
x[4] = "o"

print(x.lista)
```

```python
class RaizCuadrada:
    def __init__(self, a):
        self.a =a 
        self.n = len(self.a)

    def __iter__(self, ):
        for i in range(self.n-1):
            for j in range(self.n-1):
                if(self.a[j]>self.a[j+1]):
                    aux=self.a[j]
                    self.a[j]=self.a[j+1]
                    self.a[j+1]= aux
        return RecorrerInteracion(self.a)
    
class RecorrerInteracion:
    def __init__(self, a):
        self.a=a
        self.longitud=len(self.a)-1
        self.i=-1

    def __next__(self,):
        if self.i==self.longitud:
            raise StopIteration
        
        self.i+=1
        return self.a[self.i]**0.5


a=[81, 16, 64, 9]
x = RaizCuadrada(a)
p = iter(x)
print(next(p), next(p), next(p), next(p))
```

### "__ slots __ " + setattr + getattr

```python
class Limites: 
     __slots__ =["edad", "sueldo", "trabajo", "__dict__"]
     pass

x=Limites()
x.edad=4
print(x.edad)

setattr(x, 'sueldo', 340000)
print(getattr(x, 'edad'))
print(getattr(x, 'sueldo'))
```

### "__ mul __ " 

```python
class Numbers: 
    def __init__(self, name: int):
        self.num = name

    def __mul__(self, other: int):
        return self.num * other


result = Numbers(2)
print(result * 2)
```
